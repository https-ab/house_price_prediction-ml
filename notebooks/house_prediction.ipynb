```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Cleaning

```python
df = df.drop(columns=["id", "date"])
df.head()
```

```python
df.columns
```

Creating features & target

```python
X = df.drop("price", axis=1)
Y = df["price"]
```

```python
print(X.shape)
print(Y.shape)
```

Splitting dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#verification
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

LINEAR REGRESSION

Training Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

Making Predictions

```python
predictions_lr = lr.predict(X_test)
```

```python
print(predictions_lr[:10])
```

Evaluating Model

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
lr_mae = mean_absolute_error(Y_test, predictions_lr)
lr_mse = mean_squared_error(Y_test, predictions_lr)
lr_r2 = r2_score(Y_test, predictions_lr)

print("MAE :", lr_mae)
print("MSE :", lr_mse)
print("R2 Score :", lr_r2)
```

Visual Predictions

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(Y_test, predictions_lr, alpha=0.5)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")
plt.show()
```

RANDOM FOREST REGRESSOR

Training Model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rfr.fit(X_train, Y_train)
```

Making Predictions

```python
predictions_rfr = rfr.predict(X_test)
```

Evaluating Model

```python
rfr_mae = mean_absolute_error(Y_test, predictions_rfr)
rfr_mse = mean_squared_error(Y_test, predictions_rfr)
rfr_r2 = r2_score(Y_test, predictions_rfr)

print("MAE :", rfr_mae)
print("MSE :", rfr_mse)
print("R2 Score :", rfr_r2)
```

Visual Prediction

```python
plt.figure(figsize=(8, 6))

plt.scatter(Y_test, predictions_rfr, alpha=0.5)

plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices (Random Forest)")

plt.show()
```

Compare Both Models

```python
import pandas as pd

comparison = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest Regressor"],
    "MAE": [lr_mae, rfr_mae],
    "MSE": [lr_mse, rfr_mse],
    "R² Score": [lr_r2, rfr_r2]
})
comparison
```

```python
plt.figure(figsize=(6,4))

plt.bar(comparison["Model"], comparison["R² Score"])

plt.title("Model Comparison (R² Score)")
plt.ylabel("R² Score")

plt.show()
```

Linear Regression is simple, fast, and easy to interpret, making it a good baseline.
Random Forest Regressor can capture complex, non-linear relationships in the data and often achieves a higher R² score on this dataset.
